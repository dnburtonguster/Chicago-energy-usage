```python
#setup train and test folds
from sklearn.cross_validation import train_test_split

#engineer kwh + therms
#total_energy = df[df['TOTAL THERMS'] + df['TOTAL KWH']]

X = df[[
'TOTAL POPULATION',
'TOTAL UNITS',
'AVERAGE STORIES',
'AVERAGE BUILDING AGE',
'AVERAGE HOUSESIZE',
'OCCUPIED UNITS',
'OCCUPIED UNITS PERCENTAGE',
'RENTER-OCCUPIED HOUSING UNITS',
'RENTER-OCCUPIED HOUSING PERCENTAGE',
'OCCUPIED HOUSING UNITS',
'Commercial',
'Industrial',
'Residential',
'Commercial_sub',
'Industrial_sub',
'Multi 7+',
'Multi < 7',
'Municipal',
'Single Family',
'ELECTRICITY ACCOUNTS',
'GAS ACCOUNTS'
]]

# print X[:5]
# X = [val if "less" not in val else 1 for val in X]
Y = df['TOTAL THERMS']
#print X[:5]
# print 'now y'
# print Y[:5]

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.95, random_state=4444)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression().fit(X_train, Y_train)
Y_test_pred = model.predict(X_test)
```

```python
logit_accuracy = accuracy_score(Y_test, Y_test_pred)
print 'Logistic Regression accuracy: %.9f' % logit_accuracy
```

```python
import statsmodels.formula.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std

def train_model(dataframe, explanatory_vars, outcome_var):
    X = dataframe[explanatory_vars]
    Y = dataframe[outcome_var]
    model = sm.OLS(Y, X).fit()
    return model

lin_model = train_model(df, ['Commercial',
'Industrial',
'Residential',
'Commercial_sub',
'Industrial_sub',
'Multi 7+',
'Multi < 7',
'Municipal',
'Single Family',
'ELECTRICITY ACCOUNTS',
'GAS ACCOUNTS'], 'TOTAL THERMS')

#loglin_model = train_model(df, ['Budget', 'Ones'], 'LogGross')
```

```python
#dir(lin_model)
lin_model.summary()
```

```python
#compare model w/actual outcomes
def plot_predictions(dataframe, explanatory_vars, outcome_var, model):
    X = dataframe[explanatory_vars]
    # plot the predictions 
    plt.plot(dataframe[outcome_var], model.predict(X), 'go')
    # plot the 1:1 actual outcome line
    plt.plot(dataframe[outcome_var], dataframe[outcome_var], 'b')
    # axis labels
    plt.xlabel('Actual %s' % outcome_var)
    plt.ylabel('Predicted %s' % outcome_var)
```

```python
plot_predictions(df, ['Commercial',
'Industrial',
'Residential',
'Commercial_sub',
'Industrial_sub',
'Multi 7+',
'Multi < 7',
'Municipal',
'Single Family',
'ELECTRICITY ACCOUNTS',
'GAS ACCOUNTS'], 'TOTAL THERMS', lin_model)
```
